### What are stopwords?
Words such as the, a , an, of , their, them - these might not play a very important role in any kind of tasks we want to do with the text; but some of the words such as neither, not might have an important role to play. So, we basically find which all words we can remove without impacting any of our further steps

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
```

```python
nltk.download('stopwords')
```

```python
stopwords.words('English')
```

This may also depend on the data - in that you can also define your own stopwords

```python
stopwords.words('French')
```

```python
stopwords.words('Arabic')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(speech)
```

```python
sentences
```

```python
type(sentences)
```

```python
## Apply stopwords removal, followed by stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words("English"))]
    sentences[i] = ' '.join(words) #basically we are converting all the list of stemmed words into sentences
```

```python
sentences
```

You can see this stemming does not look that great! So, what do we do now, we have already seen earlier which Stemming is better than Porter stemming - let us use Snowball stemmer here

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer("english")
```

```python
## Apply stopwords removal, followed by Snowball stemming

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words("English"))]
    sentences[i] = ' '.join(words) #basically we are converting all the list of stemmed words into sentences
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
## Apply stopwords removal, followed by Lemmatization

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word, pos = 'v') for word in words if word not in set(stopwords.words("English"))]
    sentences[i] = ' '.join(words) #basically we are converting all the list of stemmed words into sentences
```

```python
sentences
```

```python
## Apply stopwords removal, followed by lowering sentences and Lemmatization

for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word.lower(), pos = 'v') for word in words if word not in set(stopwords.words("English"))]
    sentences[i] = ' '.join(words) #basically we are converting all the list of stemmed words into sentences
```
